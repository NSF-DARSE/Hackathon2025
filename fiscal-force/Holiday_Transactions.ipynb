```python
import bigframes.pandas as bf
```

# Ingest Data

```python
df = bf.read_gbq_table("jpmc1-468719.de_ds1.checkbookdata")
```

## Define our range of holidays

For this, we use the `holidays` library.

```python
import holidays

us_holidays = holidays.US(years=range(2020, 2026))
```

Ensure that the `CHECK_DATE` column is queriable from pandas

```python
df['CHECK_DATE'] = bf.to_datetime(df['CHECK_DATE'])
```

## Make new holiday columns

```python
df['IS_HOLIDAY'] = df['CHECK_DATE'].dt.date.isin(us_holidays.keys())
df['HOLIDAY_NAME'] = df['CHECK_DATE'].dt.date.map(us_holidays)
```

## Filter and save the transactions that happened on holidays

```python
holidays_txs = df[df['IS_HOLIDAY'] == True]
holidays_txs
```

## Export the data

```python
holidays_txs.to_gbq('jpmc1-468719.de_ds1.holiday_transactions')
```
