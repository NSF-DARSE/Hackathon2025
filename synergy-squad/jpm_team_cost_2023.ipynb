```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('de_checkbook.csv')
```

```python
df.head()
```

```python
df2= df[["FISCAL YEAR","DEPARTMENT","ACCOUNT_DESCR","VENDOR","AMOUNT"]]
df2 = df2[(df2['ACCOUNT_DESCR'] == 'OFFICE SUPPLIES') & (df2['FISCAL YEAR'] == 2023)]
df2
```

```python
result = df2.groupby('DEPARTMENT')['AMOUNT'].sum().reset_index()
result['EMPLOYEE COUNT']= 0
result.loc[result['DEPARTMENT'] == 'DEPT OF EDUCATION', 'EMPLOYEE COUNT'] = 280
result.loc[result['DEPARTMENT'] == 'DELAWARE NATIONAL GUARD', 'EMPLOYEE COUNT'] = 120
result.loc[result['DEPARTMENT'] == 'DEPT OF AGRICULTURE', 'EMPLOYEE COUNT'] = 126
result.loc[result['DEPARTMENT'] == 'DEPT OF LABOR', 'EMPLOYEE COUNT'] = 372
result.loc[result['DEPARTMENT'] == 'DEPT OF TECHNOLOGY & INFORMATI', 'EMPLOYEE COUNT'] = 309
result.loc[result['DEPARTMENT'] == 'DEPT OF FINANCE', 'EMPLOYEE COUNT'] = 260
result.head(30)
```

```python
result['COST PER EMPLOYEE'] = np.where(
    result['EMPLOYEE COUNT'] == 0,
    0,
    result['AMOUNT'] / result['EMPLOYEE COUNT']
)
result['COST PER EMPLOYEE'] = result['COST PER EMPLOYEE'].round(2)
result
```
