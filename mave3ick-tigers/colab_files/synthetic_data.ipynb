using just random oversampling

```python
import pandas as pd

LABEL = "QuoteStatus"
TARGET_TOTAL = 4000
PER_CLASS = TARGET_TOTAL // 2  # 2000 per class

def oversample_equal(df, label=LABEL, per_class=PER_CLASS, random_state=42):
    result_parts = []
    for cls in df[label].unique():
        cls_df = df[df[label] == cls]
        if len(cls_df) >= per_class:
            sampled = cls_df.sample(n=per_class, replace=False, random_state=random_state)
        else:
            sampled = cls_df.sample(n=per_class, replace=True, random_state=random_state)
        result_parts.append(sampled)
    return pd.concat(result_parts, ignore_index=True)

train_synthetic = oversample_equal(data_available)

# Shuffle rows
train_synthetic = train_synthetic.sample(frac=1.0, random_state=42).reset_index(drop=True)

# Check counts
print(train_synthetic[LABEL].value_counts())
print(train_synthetic.shape)

train_synthetic.to_excel("train_synthetic_4000.xlsx", index=False)
```
